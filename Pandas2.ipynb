#iloc


gets rows(or columns) at particular positions in the index(so it only takes integers)

```python
#Let us get a subset which consists of first 8 rows and first 4 columns
df.iloc[:5,-3:]
```

#loc

It gets rows(or columns) with particular labels from the index

```python
df.loc[50:60]
```

```python
df[0:10]
```
